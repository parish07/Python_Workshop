# Scipy Curve Fitting Example

```python
import numpy as np
import scipy.optimize as opt
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#load the text file downloaded from respository
data = np.loadtxt(r"C:\Users\santosh.d\Desktop\test_2.txt")
data.shape
```

```python
#Assign first colloumn to x and 2nd colloumn to y
x=data[0,:]
y=data[1,:]
```

```python
#Curve Fit Model Function for Linear fit like y=m*x+c
def func(x,a,b):
    return(a*x+b)   
```

```python
# Call the curve fit function, which returns the optimize parameters 
param_opt,Pcov=opt.curve_fit(func, x, y)
```

```python
y_lin = func(x,*param_opt)
plt.figure(figsize=(10,6))
plt.plot(x,y,'o')
plt.plot(x, y_lin, '-')
#plt.plot(x, y_lin, '-',label='Linear fit: a=%f, b=%f' % tuple(param_opt))
#plt.xlabel('X Values')
#plt.xlabel('Y Values')
#plt.legend()
#plt.show()
```

```python
#Curve Fit Model Function for Linear fit like y=a*x*x+ b*x + c
def func2(x,a,b,c):
    return(a*x*x+b*x+c)

param_opt1,Pcov1=opt.curve_fit(func2, x, y)
```

```python
#plt.plot(x,y,'o')
#
#plt.plot(x, y_quad, '-',label='Polynomial fit (Order 2): a=%f, b=%f, c=%f' % tuple(param_opt1))

#plt.legend()
#plt.show()
y_quad = func2(x,*param_opt1)

#plt.plot(x,y_quad)

plt.figure(figsize=(10,6))
plt.plot(x,y,'o')
plt.plot(x, y_quad, '-')
plt.xlabel('X Values')
plt.ylabel('Y Values')
```

```python
#Curve Fit Model Function for Linear fit 4 Point Logistic function
def func4pl(x, a, b, c, d):
    return a / (1. + np.exp(-c * (x - d))) + b

param_opt2,Pcov2=opt.curve_fit(func4pl, x, y)
```

```python
y_4pl = func4pl(x,*param_opt2)
plt.figure(figsize=(10,6))
plt.plot(x,y,'o')
plt.plot(x, y_4pl, '-')
plt.plot(x, y_4pl, '-',label='4 Point Logistic Regression: a=%f, b=%f, c=%f, d%f' % tuple(param_opt2))
plt.xlabel('X Values')
plt.xlabel('Y Values')
plt.legend()
plt.show()
```
